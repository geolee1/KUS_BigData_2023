# 데이터의 운영(Pandas)

## 시리즈(Series)

### 시리즈의 생성 : 리스트

```python
import pandas as pd
# 리스트로 Series 생성
list_data = ['Dog', 'Cat', 'Tiger', 'Lion', 'Monkey']
sd1 = pd.Series(list_data, index=['0', '1', '2', '3', '4'])
# 데이터 확인
sd1
```

### 시리즈의 생성 : 딕셔너리

```python
import pandas as pd
# 딕셔너리로 Series 생성
dict_data = {'a':1, 'b':2, 'c':3}
sd2 = pd.Series(dict_data)
# 데이터 확인
sd2
```

### 시리즈의 생성 : 튜플

```python
import pandas as pd
# 튜플로 Series 생성
tup_data = ('이순신', '1991-03-15', '남')
sd3 = pd.Series(tup_data, index=['이름', '생년월일', '성별'])
# 데이터 확인
sd3
```

## 데이터프레임(DataFrame)

### 데이터프레임의 생성 : 리스트

```python
# 라이브러리 불러오기
import pandas as pd

# 리스트로 데이터프레임 생성 - 중첩된 리스트 사용
df1 = pd.DataFrame([['Dog', '7'], ['Cat', '9'], ['Tiger', '2'], ['Lion', '3'], ['Monkey', '1']],
                   index=['0', '1', '2', '3', '4'], 
                   columns=['동물', '나이'])

# 타입 확인
print(type(df1))
# 데이터 확인
print(df1)
```

### 데이터프레임의 생성 : 딕셔너리

```python
# 라이브러리 불러오기
import pandas as pd

# 딕셔너리로 데이터프레임 생성 - 딕셔너리안의 리시트
dict_data = {'동물':['Dog', 'Cat', 'Tiger', 'Lion', 'Monkey'],'나이':[7, 9, 2, 3 ,1]}
df2 = pd.DataFrame(dict_data)

# 타입 확인
print(type(df2))
# 데이터 확인
print(df2)
```
